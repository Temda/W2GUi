## Modules

Modules are basically files which contain definitions that provide helpful functionality. Theoretically, all of you could write any program you wanted to write on your own, without any modules or outside help. But why reinvent the wheel if you don't have to? This is the general gist behind modules. You all have experience using modules such as tkinter and csv. These modules are part of a library of standard modules that came packaged with your install which the developers have deemed useful or necessary for your daily python usage.





## Importing modules in Python

There are three different ways to import models in Python

1. `import model_name`
2. `import model_name as new_name`
3. `from model_name import *`

** We have already prepared functions in `module1.py`. Click the file for more information. **

**Let's try!**

## Example 1 : Calling functions from other file using import

```python
import module1
```

After you imported `module1.py`, you can call functions, which are defined in the file no need to reinvent by using `module1`

```python
print("------- Get started -----")
module1.display()
module1.displayname()
print("-"*30)
name = module1.getname()
print("Hi ",name)
print("-"*30)

```

## Example2 : Calling functions from other file using import .. As ..

```python
import module1 as me
```

This example is similar to above example **BUT** adds alias `me`.

```python
print("------- Get started -----")
me.display()
me.displayname()
print("-"*30)
name = me.getname()
print("Hi ",name)
print("-"*30)
```

## Example3 : Calling function from other file using From .. Import..

One feature of this importing, you can import some functions as you need and you can also import all by using `*`

```python
from module1 import *
```

**Notice** you can call functions directly without alias.

```python
print("------- Get started -----")
display()
displayname()
print("-"*30)
name = getname()
print("Hi ",name)
print("-"*30)
```
